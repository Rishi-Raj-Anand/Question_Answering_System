```python
from transformers import pipeline
```

```python
# Load the pre-trained Question Answering model
print("Loading model...")
qa_pipeline = pipeline(
    "question-answering", 
    model="distilbert-base-cased-distilled-squad"
)
print("Model loaded successfully!")
```

```python
# Defining context 
context = """
The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France.
It is named after the engineer Gustave Eiffel, whose company designed and built the tower.
Constructed from 1887 to 1889 as the centerpiece of the 1889 World's Fair, it was
initially criticized by some of France's leading artists and intellectuals, but it has
become a global cultural icon of France and one of the most recognizable structures in the world.
The tower is 330 metres (1,083 ft) tall.
"""
```

```python

# Define the questions you want to ask
questions = [
    "Where is the Eiffel Tower located?",
    "Who designed the Eiffel Tower?",
    "How tall is the Eiffel Tower?",
    "When was the Eiffel Tower built?",
    "Who criticized it?"
]
```

```python
print("\n--- Answering Questions ---")

# Run the questions through the pipeline
for question in questions:
    # The pipeline takes the question and context
    result = qa_pipeline(question=question, context=context)
    
    # Print the deliverable: Question and System Response
    print(f"\nQ: {question}")
    print(f"A: {result['answer']} (Confidence: {result['score']:.4f})")
```
